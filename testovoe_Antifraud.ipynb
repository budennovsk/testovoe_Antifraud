```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Отбираем только числовые колонки
df_numeric = df_behavior_task.select_dtypes(include=['number'])

# Вычисление корреляционной матрицы для числовых колонок
correlation_matrix = df_numeric.corr()

# Фильтруем корреляции, чтобы оставить только те, которые больше 0.30 или меньше -0.30
filtered_correlation = correlation_matrix[(correlation_matrix > 0.30) | (correlation_matrix < -0.30)]



# Построение тепловой карты
plt.figure(figsize=(10, 6))
sns.heatmap(filtered_correlation, annot=True, cmap='coolwarm', vmin=-1, vmax=1)

# Настроим график
plt.title('Корреляционная матрица признаков (без object-колонок)')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant


# Отбираем только числовые колонки
df_numeric_1 = df_behavior_task.select_dtypes(include=['number'])
# Заполняем пропуски на нули
df_numeric_1 = df_numeric_1.fillna(0)

# Добавляем константу для расчета VIF
df_with_const = add_constant(df_numeric_1)

# Рассчитываем VIF для каждого признака
vif_data = pd.DataFrame()
vif_data["feature"] = df_with_const.columns
vif_data["VIF"] = [variance_inflation_factor(df_with_const.values, i) for i in range(df_with_const.shape[1])]

# Печатаем VIF для каждого признака
print("Факторы инфляции дисперсии (VIF):")
print(vif_data)

# Фильтруем признаки с VIF > 5 (или можно установить порог > 10 для более строгой проверки)
high_vif = vif_data[vif_data["VIF"] > 5]
print("\nПризнаки с высоким VIF (> 5):")
print(high_vif)
```
